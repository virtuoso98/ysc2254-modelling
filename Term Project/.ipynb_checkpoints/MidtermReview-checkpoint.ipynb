```python
import numpy as np
from scipy.optimize import minimize
```

```python
def objective(x):
    var = 0
    mean_util = 0
    for i in range(35):
        mean_util += x[i]
    mean_util /= 7
    for j in range(7):
        var_day = 0
        for k in range(1, 6):
            print(x[(j * 5) + k - 1])
            var_day += x[(j * 5) + k - 1]
        var += (var_day - mean_util) ** 2
    return var / 7
        
```

```python
def constraint1(x):
    res = 0
    for i in range(35):
        res += x[i] ** 2
    return res
```

```python
x0 = np.zeros(35)
b = (1.0, 5.0)
bnds = tuple(b for i in range(35))
con1 = {'type': 'ineq', 'fun': constraint1}
cons = ([con1])
solution = minimize(objective,x0,method='SLSQP',\
                    bounds=bnds,constraints=cons)

x = solution.x
```
